## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

1.Data augmentations like resizing, cropping, rotating, flipping, and adjusting brightness or contrast are commonly used techniques in training deep learning models, especially for computer vision tasks. Their significance lies in several key benefits they bring to the training process:

Increased Dataset Size: Augmentations effectively increase the effective size of your dataset. This is crucial, especially when dealing with limited amounts of labeled data. More diverse data helps prevent overfitting and improves generalization.

Regularization: Augmentations act as a form of regularization by adding noise or variations to the training data. This helps the model generalize better to unseen examples, reducing the risk of memorizing training samples.

Improved Robustness: Models trained with augmented data tend to be more robust to variations and distortions present in real-world data. For example, a model trained on augmented images may better handle variations in lighting, perspective, or object sizes.

Domain Adaptation: Augmentations can help bridge domain gaps. For instance, if your training data comes from one source (e.g., professional photos) but you expect the model to perform well on images from another source (e.g., smartphone photos), augmentations that mimic the characteristics of the target domain can be beneficial.

Reduced Bias: Augmentations can help reduce biases in the training data by introducing diverse examples. This is particularly important for tasks where the training data might be skewed or biased toward certain classes or conditions.

Data Balancing: Augmentations can help balance class distributions in the dataset. For example, if certain classes are underrepresented, augmenting the existing samples can create a more balanced training set.

Improved Performance: In many cases, models trained with augmented data achieve better performance metrics such as accuracy, precision, and recall on unseen data compared to models trained without augmentation.

If you train a ResNet model from scratch without using pretraining (i.e., setting pretrained=False), the performance can vary significantly depending on factors such as the dataset size, complexity of the task, available compute resources, and training hyperparameters. Here are some general expectations and considerations:

Training Time: Training a deep neural network like ResNet from scratch can be computationally expensive and time-consuming, especially if you have a large dataset and complex model architecture. It may require substantial computational resources, including GPU accelerators, to train efficiently.

Data Requirements: Deep learning models trained from scratch typically require a larger amount of labeled data to achieve good performance compared to models initialized with pretrained weights. If your dataset is small, training from scratch might lead to overfitting or suboptimal generalization.

Model Initialization: When training from scratch, the model's weights are initialized randomly. This means the model starts with no prior knowledge about the task and must learn meaningful representations solely from the training data. In contrast, pretrained models already have learned features from a different task or dataset, providing a helpful starting point.

Convergence Speed: Training from scratch may require more epochs for the model to converge and achieve good performance compared to starting with pretrained weights. It may involve experimenting with different learning rates, optimizers, and regularization techniques to find the optimal training configuration.

Performance Comparison: The performance of a ResNet model trained from scratch can vary depending on the specific dataset and task. It's generally expected that using pretrained weights as a starting point (transfer learning) leads to faster convergence and better generalization, especially when the pretrained model was trained on a large, diverse dataset (e.g., ImageNet).

Evaluation Metrics: The performance of the model can be evaluated using various metrics such as accuracy, precision, recall, F1 score, and loss function values. Comparing these metrics between models trained with and without pretrained weights can provide insights into their relative performance.

To evaluate the performance of a ResNet model trained from scratch, you would follow a typical deep learning workflow, including data preprocessing, model definition, training with appropriate loss functions and optimizers, and evaluation using validation/test data. Monitoring training/validation metrics over epochs and comparing against pretrained models or baseline models can help assess the impact of training from scratch.
